```python
from random import randint

from rx import generate


fails = 0
totalTests = 1_000_000
print(f"Testing: {totalTests} times")

def generateScenraio():
    boxes = {}
    numbers = list(range(100))
    prisoners = numbers.copy()

    for i in range(100):
        rnd = randint(0,len(numbers) - 1)
        boxes[i] = numbers[rnd]
        numbers.pop(rnd)
    return boxes, numbers, prisoners

def runTest():
    boxes, numbers, prisoners = generateScenraio()
        
    for i in prisoners:
        maxGuesses = 50
        currentGuess = i
        while (maxGuesses > 0 and boxes[currentGuess] != i):
            currentGuess = boxes[currentGuess]
            maxGuesses = maxGuesses - 1
                
        if(maxGuesses == 0):
            return 1
    return 0
```

```python
#Run
for i in range(totalTests):
    fails += runTest()

print(f"Success {totalTests - fails}/{totalTests}")
```
